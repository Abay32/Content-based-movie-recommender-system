## Recommender system

Basically there are two types of recommender systems 
1. content based fitering 
2. coaborative based fitering 


```python
import numpy as np
import pandas as pd
```

```python
col_name =  ['user_id', 'item_id', 'rating', 'timestamp'] # replace column names 
df = pd.read_csv('u.data', sep='\t', names= col_name)
```

```python
df.head()
```

```python
Movie_titles = pd.read_csv('Movie_Id_Titles.csv')
```

```python
Movie_titles.head()
```

```python
# Merge the the two data sets
df = pd.merge(df, Movie_titles, on='item_id')
```

```python
df.head(100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
# pot the histogram based on the numer of ratings
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=60)
```

```python
sns.jointplot(x='rating', y='num of ratings', data=ratings, alpha = 0.5)
```

```python
## Recommending simiar movies, 
```

```python
moviemat = df.pivot_table(index='user_id', columns = 'title', values = 'rating')
```

```python
moviemat.head()
```

```python
ratings.sort_values('num of ratings', ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns = ['Correlation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation', ascending=False)
```

```python
# Exercise do the same for liarliar 
corr_liar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'] )
corr_liar.dropna(inplace=True)
corr_liar.head()
```

```python
corr_liar = corr_liar.join(ratings['num of ratings'])
corr_liar.head()
```

```python
corr_liar[corr_liar['num of ratings']>100].sort_values('Correlation', ascending=False).head()
```
